### **Botpress**
[Сайт](https://botpress.com/docs)

[Репозиторий](https://github.com/botpress/botpress)

Botpress - это стандартный стек разработчика для создания, запуска и улучшения приложений Conversational-AI. Botpress, основанный на понимании естественного языка, API обмена сообщениями и полнофункциональной студии, позволяет разработчикам по всему миру создавать замечательных чат-ботов.

**По умолчанию Botpress включает в себя:**
- Панель администратора для управления и мониторинга ваших чат-ботов
- Conversation Studio для разработки беседы, управления контентом, пользовательской интеграции кода
- Простая интеграция с каналами обмена сообщениями (Messenger, WhatsApp, Slack, Teams, Webchat, Telegram, SMS и др.)
- Понимание естественного языка

### **Botfront**

[Сайт](https://botfront.io/)

[Репозиторий](https://github.com/botfront)

Botfront - это полнофункциональная диалоговая платформа (платформа корпоративного уровня), построенная на основе Rasa с интуитивно понятными интерфейсами для создания контекстно-зависимых виртуальных помощников.

Botfront предоставляет все стандартные блоки Rasa (правила, истории, формы) в удобном и интуитивно понятном пользовательском интерфейсе. Важно отметить, что для поддержки основных функций Botfront, таких как пользовательский интерфейс блок-схемы форм и многоязыковую поддержку.

Botfront требуется разветвленная версия Rasa. Botfront не будет работать с канонической Rasa, но вы всегда можете экспортировать свои проекты из Botfront, если хотите использовать его со стандартной средой выполнения Rasa.

[Документация](https://botfront.io/docs/rasa/getting-started/)

### **Rasa Ui**

[Репозиторий](https://github.com/paschmann/rasa-ui)

Rasa UI - это веб-приложение, созданное для Rasa. Пользовательский интерфейс Rasa предоставляет веб-приложение для быстрого и простого создания и управления ботами, компонентами NLU (регулярное выражение, примеры, сущности, намерения и т.д.) и основными компонентами (истории, действия, ответы и т.д.) Он также предоставляет некоторые удобные функции для Rasa, такие как обучение и загрузка ваших моделей, мониторинг использования или просмотр журналов.

**Функции**
- Пользовательский интерфейс для создания обучающих данных и управления ими - Примеры, Намерения, Сущности, Синонимы, Регулярное выражение, Истории, Действия, Ответы
- Управление несколькими ботами из одного пользовательского интерфейса / экземпляра Rasa UI.
- Создавание, управление и загрузка разных версий моделей для тестирования и оптимизации.
- Запросы журнала для отслеживания использования, истории и улучшений ваших моделей
- Легко выполнять анализ намерений с использованием различных моделей
- Данные хранятся в базе данных SQLite для резервного копирования / совместного использования
- Может использоваться с бэкэндом Rasa или без него для управления данными обучения
